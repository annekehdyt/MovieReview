```python
import os
import sys
import csv
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
#parsing negative phrases
with open('Negative.tsv') as csvfile:
    reader=csv.reader(csvfile, delimiter='\t')
    collect={} # key:phrase value:number of times appeared in the collection
    for row in reader:
        for i in row:
            if i is not "":
                if i not in collect:
                    collect[i]=1
                else:
                    collect[i]+=1  
#     #printing stats
#     for count in range(0,2):
#         print('\n')
#         print("Phrases that appear ",count+1," times:")
#         print('\n')
#         for z in collect:
#             if(collect[z]==count+1):
#                 print(z)
```

```python
#Lowercase negative phrases
NegphraseL=[]
for x in collect: 
    NegphraseL.append(x.lower())
newList=[]
for y in NegphraseL:
    if y not in newList:
        newList.append(y)
print(newList)
```

```python
directory = os.path.normpath("C:/Users/meghu/Desktop/train/neg")
counter=0
doc=[]
for subdir, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith(".txt"):
            f=open(os.path.join(subdir, file),'r')
            try:
                a = f.read() 
                doc.append(a)
            except:
                pass
            counter+=1
            #print(counter)
            f.close()
#print(doc[1])
```

```python
vect=CountVectorizer(vocabulary=newList, lowercase=True, ngram_range=(2,20))
```

```python
X=vect.fit_transform(doc)
```

```python
np.sum(X, axis=0)
```

```python
voc=vect.vocabulary_
```

```python
print(voc)
```

```python
#prints number of positive reviews that contain the positive manual phrases
for y in voc:
    print(y,": ",voc[y])
```
